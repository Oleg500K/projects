### Проверка расчета эффективности обогащения

Проверим, что эффективность обогащения рассчитана правильно. Вычислим ее на обучающей выборке для признака rougher.output.recovery.


Recovery = (C(F-T))*100%/F(C-T)

C — доля золота в концентрате после флотации rougher.output.concentrate_au;
F — доля золота концентрате до флотации rougher.input.feed_au;
T — доля золота в отвальных хвостах после флотации rougher.output.tail_au.
Recovery - rougher.output.recovery_calc



```python
# осуществим расчет значения столбца 'rougher.output.recovery_calc'
df_train['rougher.output.recovery_calc'] = (
    df_train['rougher.output.concentrate_au'] * (df_train['rougher.input.feed_au'] - df_train['rougher.output.tail_au'])
) * 100 / (
    df_train['rougher.input.feed_au'] * (df_train['rougher.output.concentrate_au'] - df_train['rougher.output.tail_au'])
)

print(df_train[['rougher.output.recovery_calc', 'rougher.output.recovery']])
```

```python
# осуществим расчет MAE
mae = mean_absolute_error(df_train["rougher.output.recovery_calc"], df_train["rougher.output.recovery"])

print('Ошибка MAE расчета коэффициента обогащения rougher.output.recovery:')
print(f'MAE: {mae:.15f}')
```

Результат ошибки MAE (Mean Absolute Error) составляет 0.000000000000011. Это означает, что средняя абсолютная разница между рассчитанными нами значениями rougher.output.recovery_calc и имеющимися значениями rougher.output.recovery очень мала. В данном случае, значение MAE практически равно нулю, что показывает факт что коэффициент обогащения рассчитан правильно.

Сразу заложим в код возможность оценить интервалы перерывов в записи.

```python
# создадим список для хранения периодов разрывов данных
discontinuity_periods = []

# инициализируем счетчик для подсчета нарушений в записи данных длительностью 2 часа
discontinuity_count_2_hours = 0

# инициализируем переменную для хранения предыдущей даты
prev_date = None

# пройдем циклом по каждой дате в столбце 'date'
for date in df_full['date']:
     # проверим, есть ли предыдущая дата и разница во времени не равна 1 часу (3600 секунд)
    if prev_date is not None and (date - prev_date).total_seconds() != 3600:

        # вычислим длительность разрыва данных в часах
        duration_hours = (date - prev_date).total_seconds() / 3600
        if duration_hours == 2:
            discontinuity_count_2_hours += 1

        # добавим информацию о разрыве в список discontinuity_periods
        discontinuity_periods.append({'Начало': prev_date, 'Конец': date, 'Простой, ч': duration_hours})

    # обновим значение предыдущей даты
    prev_date = date

print("Количество превышения периода записи данных в df_full:", len(discontinuity_periods))
print("Количество превышения периода записи данных (продолжительностью = 2 ч) в df_full:", discontinuity_count_2_hours)

# выведем информацию о временных периодах сбоя записи данных
if discontinuity_periods:
    discontinuity_periods_df = pd.DataFrame(discontinuity_periods)
    sorted_df = discontinuity_periods_df.sort_values(by='Простой, ч', ascending=False)
    display(f'Временные периоды сбоя записи данных в df_full:')
    print(sorted_df.head(15))
    print(sorted_df.tail(5))
else:
    print(f'\nСбоев в записи данных в df_full не обнаружено.')
```

```python
# создадим список продолжительностей простоев в часах
durations = [period['Простой, ч'] for period in discontinuity_periods]

# установим размеры фигуры
plt.figure(figsize=(12, 6))

# построим гистограммы
plt.hist(durations, bins=50)
plt.xlabel('Продолжительность простоя, ч')
plt.ylabel('Количество периодов')
plt.title('Гистограмма простоев записи данных в df_full')
plt.grid(True)

plt.show()
```

Выводы: гипотеза подтверждена. Действительно данные имеют запись с периодом 1 час, но есть большое количество прерываний записи данных. Максимальное прерывание составляет 195 часов. Естественно предположить, что после каждого перерыва, особенно если он велик, часть данных скорее всего будет отсутствовать, так как составу необходимо физически дойти из точки загрузки руды до точки, где получается итоговый состав, из которого будут извлекать золото.

Посмотрим распределение рабочего времени, построим гистограмму.

```python
# cоздадим пустой список для хранения продолжительностей непрерывных отрезков времени
continuous_periods = []
prev_date = None  # инициализируем переменную для хранения предыдущей даты

# инициализируем переменную для хранения начала текущего непрерывного отрезка времени
current_period_start = None

# инициализируем переменную для хранения продолжительности текущего отрезка времени
current_period_duration = 0

# проходим по каждой дате в столбце 'date'
for date in df_full['date']:
    if prev_date is None:
        current_period_start = date # если это первая дата, сохраняем ее как начало текущего отрезка
        current_period_duration = 1 # установим продолжительность текущего отрезка равной 1
    else:
        if (date - prev_date).total_seconds() == 3600:
            # если разница между текущей и предыдущей датами равна 1 часу, увеличиваем продолжительность текущего отрезка
            current_period_duration += 1
        else:
            # если разница больше 1 часа, добавляем продолжительность текущего отрезка в список
            continuous_periods.append(current_period_duration)
            current_period_start = date # установим новое начало текущего отрезка
            current_period_duration = 1 # установим продолжительность нового отрезка равной 1

    prev_date = date # сохраним текущую дату в качестве предыдущей для следующей итерации

continuous_periods.append(current_period_duration)# добавим продолжительность последнего отрезка в список

# выведем информацию о непрерывных отрезках записи времени
continuous_periods_df = pd.DataFrame(continuous_periods, columns=['Продолжительность, ч']) # создадим датафрейм с продолжительностями отрезков

# отсортируем продолжительности по количеству
sorted_periods_df = continuous_periods_df['Продолжительность, ч'].value_counts().reset_index()
sorted_periods_df.columns = ['Продолжительность, ч', 'Количество'] # переименуем столбцы
sorted_periods_df = sorted_periods_df.sort_values(by='Количество', ascending=False) # отсортируем датафрейм по убыванию количества
print(sorted_periods_df.head(20))

plt.figure(figsize=(10, 12))

plt.subplot(2, 1, 1)
plt.hist(continuous_periods, bins=range(min(continuous_periods), max(continuous_periods) + 2), align='left', rwidth=0.8)
plt.xlabel('Продолжительность отрезка, ч')
plt.ylabel('Количество отрезков')
plt.title('Гистограмма распределения непрерывных отрезков записи времени')
plt.grid(True)

plt.tight_layout()
plt.show()
```

Судя по получившимся данным в основном оборудование работало непрерывно от 1 до 50 часов. Но так же было достаточно много длительных включений с большим временем, а так же совсем небольшое число периодов с очень продолжительными промежутками более 300 часов.

## Модель

### Функция для вычисления итогового *sMAPE*

```python
# создадим функцию sMAPE
def sMAPE(y_true, y_pred):
    error = np.mean(abs(y_true - y_pred)/((abs(y_true) + abs(y_pred))/2))*100
    return error
```

```python
# создадим функцию final_sMAPE
def final_sMAPE(sMAPE_rougher, sMAPE_final):
    return (0.25*sMAPE_rougher + 0.75*sMAPE_final)
```

```python
# объявим кастомную метрику качества
custom_score = make_scorer(sMAPE, greater_is_better=False)
```
